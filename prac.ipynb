### "0"を挿入 (Sepal.Lengthの列のみ)

```r
cat("0を挿入する行の行数を作成")
zero <- seq(1, 150, by = 10)
zero

iris2[zero,1] <- 0
iris2
```

### Sepal.Length列で"0"の行を特定

#### which()を使う。

```r
iris3 <- iris2
# Sepal.Length列で"0"の行の特定
zero.row <- iris3$Sepal.Length
zero.nrow <- which(zero.row == 0)
cat("0の値が入っている行番号")
zero.nrow
```

### Sepal.Length列の値が"0"の場合、"10"で置換する

```r
iris4 <- iris2
iris4[zero.nrow,]
```

```r
iris4[zero.nrow,1] <- 10
iris4[zero.nrow,]
iris4
```

## Speciesごとに"0"行を特定

```r
iris5 <- iris2
iris5
```

### 種のベクトルを作成

```r
species <- as.vector(unique(iris5$Species))
species
```

### 種ごとに置換する値の作成

```r
replace.num <- seq(10, 30, by = 10)
replace.num
```

### 表に整形 （マスタ用）

```r
tbl1 <- as.data.frame(cbind(species, replace.num))
name <- names(tbl1)
tbl1$replace.num <- as.numeric(as.character(tbl1$replace.num))
names(tbl1) <- name
tbl1
```

### import package

```r
library(dplyr)
```

### 種別ごとにwhich()

```r
zero.bySpe <- {}
iris.rep <- {}
for(i in species) {
    a <- dplyr::filter(iris5, Species == i)
    zero.bySpe <- append(zero.bySpe, which(a$Sepal.Length == 0))
    a[zero.bySpe,1] <- filter(tbl1, species == i)$replace.num
    iris.rep <- rbind(iris.rep, a)
    rm(a)
}
zero.bySpe
iris.rep
```
