## which()

### ある条件を満たす行列要素の位置を返す。

```r
iris2 <- iris
iris2
```

### "0"を挿入 (Sepal.Lengthの列のみ)

```r
cat("0を挿入する行の行数を作成")
zero <- seq(1, 150, by = 10)
zero

iris2[zero,1] <- 0
iris2
```

### Sepal.Length列で"0"の行を特定

#### which()を使う。

```r
iris3 <- iris2
# Sepal.Length列で"0"の行の特定
zero.row <- iris3$Sepal.Length
zero.nrow <- which(zero.row == 0)
cat("0の値が入っている行番号")
zero.nrow
```

### Sepal.Length列の値が"0"の場合、"10"で置換する

```r
iris4 <- iris2
iris4[zero.nrow,]
```

```r
iris4[zero.nrow,1] <- 10
iris4[zero.nrow,]
iris4
```
